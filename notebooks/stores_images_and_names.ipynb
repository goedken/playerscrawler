```python
import pandas as pd
import json
from nba_py import player
from nba_py import league
```

```python
with open('player_image_urls.json') as data_file:
    player_urls = json.load(data_file)
```

```python
espn_data = []
i = 0
for team in player_urls:
    player_urls = team['image_urls']
    for url in player_urls:
        player_url = url.split()[2][:-1]
        index = int(player_url.split('/')[7])
        player_name = str(player_url.split('/')[8])
        # print str(index) + ': ' + str(player_name)
        espn_data.append({'player_id': index, 'player_name': player_name})
        # print i
        i += 1
        #print index
print type(espn_data)
print len(espn_data)
```

```python
espn_df = pd.DataFrame(espn_data)
```
